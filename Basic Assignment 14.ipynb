```python
What does RGBA stand for?
RGBA stands for Red Green Blue and Alpha(transparency) in a color.Each of these component
values is an integer from 0 (none at all) to 255 (the maximum). These RGBA values are 
assigned to individual pixels; a pixel is the smallest dot of a single color the computer 
screen can show (as you can imagine, there are millions of pixels on a screen).
```

```python
3. What is a box tuple, and how does it work?
The box tuple of four integer coordinates that represent a rectangular region in an image. 
The four integers are, in order, as follows:
Left: The x-coordinate of the leftmost edge of the box.

Top: The y-coordinate of the top edge of the box.

Right: The x-coordinate of one pixel to the right of the rightmost edge of the box. 
    This integer must be greater than the left integer.

Bottom: The y-coordinate of one pixel lower than the bottom edge of the box. 
    This integer must be greater than the top integer.

A rectangle region is formed using top left corner and bottom right corner using co-ordinates given.
```
